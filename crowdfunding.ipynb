```python
import numpy as np
import pandas as pd
import seaborn as sns; 
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
data = pd.read_excel("C://Users/zhang/Downloads/crowd data.xlsx")
```

```python
data
```

```python
advlist=data['adv'].values #get array for advantage
```

```python
advlist = advlist[~pd.isnull(advlist)] #clear null data
```

```python
form = '+'.join(advlist) #create a list
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
import jieba
plt.figure(figsize=(20,10)) # enlarge the graph
text_from_file_with_apath = form
 
wordlist_after_jieba = jieba.cut(text_from_file_with_apath, cut_all = True)
wl_space_split = " ".join(wordlist_after_jieba) #text split
stopwords=set(STOPWORDS) #remove unrelated words
stopwords.add('investment')
stopwords.add('investments')
stopwords.add('investor')
stopwords.add('core')
stopwords.add('versus')
stopwords.add('one')
stopwords.add('good')
stopwords.add('plus')
stopwords.add('average')
stopwords.add('excellent')
stopwords.add('industry')
stopwords.add('million venture')
stopwords.add('investing')
stopwords.add('fund')
stopwords.add('substantial')
my_wordcloud = WordCloud(background_color='white',max_words=30,stopwords=stopwords,max_font_size=36,random_state=42).generate(wl_space_split)
 
plt.imshow(my_wordcloud)
plt.axis("off")
plt.show()
#get the wordcloud graph
```

```python
# get barplots
plt.figure(figsize=(20,10)) 
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
matplotlib.rc('font', **font)
s = data['Category'].value_counts()
ax=s.plot.bar(width=.8) 
for i, v in s.reset_index().iterrows():
    ax.text(i, v.Category + 0.2 , v.Category, color='black')
```
